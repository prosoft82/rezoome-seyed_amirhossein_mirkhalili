```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\PC\Desktop\محرومیت\data\labour_force_all_3.csv',low_memory=False)
```

```python
gender_map = {
    1: 'Male',
    2: 'Female'
}
df['Gender_Label'] = df['F2_D04'].map(gender_map)
# گروه‌بندی
grouped = df.groupby(['Gender_Label', 'deprived']).size().unstack(fill_value=0)
# تغییر اسم ستون‌ها
grouped.columns = ['Not Deprived', 'Deprived']
# درصد در هر سطر (مجموع هر سطر = ۱۰۰٪)
percent = grouped.div(grouped.sum(axis=1), axis=0) * 100
percent = percent.round(1)
# نمایش جدول نهایی
print(percent)
```

```python

gender_map = {
    1: 'Male',
    2: 'Female'
}

region_map = {
    1: 'Urban',
    2: 'rural'
}
# ۳. ایجاد ستون‌های برچسب‌دار
df['Gender_Label'] = df['F2_D04'].map(gender_map)
df['Region_Label'] = df['urban_rural'].map(region_map)

grouped = df.groupby(['Gender_Label', 'Region_Label', 'deprived']).size().unstack(fill_value=0)
grouped.columns = ['Not Deprived', 'Deprived']
percent = grouped.div(grouped.sum(axis=1), axis=0) * 100
percent = percent.round(1)
print("Gender x Region - Deprivation")
print("="*55)
print(percent)
```
