```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\PC\Desktop\محرومیت\data\labour_force_all_3.csv',low_memory=False)
```

```python
## مرحله ۵: آزمون Chi-Square

from scipy import stats

for var in ['dep_sex', 'dep_urban_rural', 'dep_age_group', 'dep_edu', 'F2_D19']:
        ct = pd.crosstab(df[var], df['Deprived'])
        chi2, p, dof, expected = stats.chi2_contingency(ct)
        v = (chi2 / (len(df) * (min(ct.shape) - 1))) ** 0.5
        print(f"{var}: χ²={chi2:.1f}, p={p:.2e}, V={v:.4f}")
```
