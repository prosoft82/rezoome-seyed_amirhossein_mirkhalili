```python
import pandas as pd

# ============================================
# دیکشنری کد استان‌ها
# ============================================

province_map = {
    0: 'Markazi',
    1: 'Gilan',
    2: 'Mazandaran',
    3: 'East Azerbaijan',
    4: 'West Azerbaijan',
    5: 'Kermanshah',
    6: 'Khuzestan',
    7: 'Fars',
    8: 'Kerman',
    9: 'Razavi Khorasan',
    10: 'Isfahan',
    11: 'Sistan and Baluchestan',
    12: 'Kurdistan',
    13: 'Hamadan',
    14: 'Chaharmahal and Bakhtiari',
    15: 'Lorestan',
    16: 'Ilam',
    17: 'Kohgiluyeh and Boyer-Ahmad',
    18: 'Bushehr',
    19: 'Zanjan',
    20: 'Semnan',
    21: 'Yazd',
    22: 'Hormozgan',
    23: 'Tehran',
    24: 'Ardabil',
    25: 'Qom',
    26: 'Qazvin',
    27: 'Golestan',
    28: 'North Khorasan',
    29: 'South Khorasan',
    30: 'Alborz'
}

# ============================================
# اضافه کردن ستون جدید
# ============================================

df['Province'] = df['province_code'].map(province_map)

# ============================================
# بررسی نتیجه
# ============================================

print(df[['province_code', 'Province']].head(10))
print(f"\nتعداد استان‌های منحصربه‌فرد: {df['Province'].nunique()}")
```

```python
# ============================================
# روش اول: با استفاده از map
# ============================================

marital_map = {
    1: 1,    # married → With spouse
    2: 0,    # biveh → Without spouse
    3: 0,    # divorced → Without spouse
    4: 0     # single → Without spouse
}

df['dep_martial_1'] = df['F2_D19'].map(marital_map)
```

```python
# ============================================
# دیکشنری نقشه
# ============================================

marital_deprivation_map = {
    1: 1,    # married → High deprivation
    2: 1,    # widowed → High deprivation
    3: 1,    # divorced → High deprivation
    4: 0     # single → Low deprivation
}

df['dep_martial_2'] = df['F2_D19'].map(marital_deprivation_map)
```

```python
# ============================================
# دیکشنری نقشه (بیوه را NaN در نظر می‌گیریم)
# ============================================

marital_spouse_map = {
    1: 1,    # married → With spouse
    2: None, # widowed → حذف (NaN)
    3: 0,    # divorced → Without spouse
    4: 0     # single → Without spouse
}

df['dep_martial_3'] = df['F2_D19'].map(marital_spouse_map)

# حذف سطرهای دارای NaN (بیوه‌ها)
df_clean = df.dropna(subset=['dep_martial_3'])

# تبدیل به int
df_clean['dep_martial_3'] = df_clean['dep_martial_3'].astype(int)

print("تعداد بیوه‌های حذف شده:", len(df) - len(df_clean))
print(df_clean[['F2_D19', 'dep_martial_3']].head(10))
```
